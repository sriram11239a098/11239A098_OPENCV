1.Read image in grayscale.

```python
from google.colab import files
uploaded = files.upload()

import cv2
from google.colab.patches import cv2_imshow

img_path = next(iter(uploaded))

# read grayscale
img = cv2.imread(img_path, 0)

cv2_imshow(img)
```

2.Apply Sobel (X and Y).

```python
sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=5)
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=5)

cv2_imshow(sobelx)
cv2_imshow(sobely)
```

3.Apply Laplacian.

```python
laplacian = cv2.Laplacian(img, cv2.CV_64F)

cv2_imshow(laplacian)
```

4.Apply Canny

```python
canny = cv2.Canny(img, 100, 200)

cv2_imshow(canny)
```

compare Sobel, Laplacian, and Canny edge detection methods.
 Difference (write this in record)  Sobel:

Detects edges in X and Y directions separately

Good for finding horizontal & vertical edges

Laplacian:

Detects edges using second derivative

Finds all directions, but more sensitive to noise

Canny:

Best edge detector

Uses multi-step process (noise removal + gradient + threshold)

Produces clear and thin edges
