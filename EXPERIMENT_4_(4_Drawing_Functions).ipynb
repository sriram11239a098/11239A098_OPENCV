1.Draw a line on an image

```python
import cv2
from google.colab.patches import cv2_imshow
from google.colab import files

img = cv2.imread(list(files.upload().keys())[0])

cv2.line(img, (50,50), (200,200), (0,255,0), 2)
cv2_imshow(img)
```

2.Draw a rectangle and circle

```python
import cv2
from google.colab.patches import cv2_imshow
from google.colab import files

img = cv2.imread(list(files.upload().keys())[0])

cv2.rectangle(img, (50,50), (200,200), (145,0,0), 2)
cv2.circle(img, (150,150), 50, (0,0,255), 2)

cv2_imshow(img)
```

3.Draw multiple geometric shapes

```python
import cv2
from google.colab.patches import cv2_imshow
from google.colab import files

img = cv2.imread(list(files.upload().keys())[0])

cv2.line(img,(10,10),(200,10),(255,0,0),2)
cv2.rectangle(img,(50,50),(150,150),(0,255,0),2)
cv2.circle(img,(200,200),40,(0,0,255),2)

cv2_imshow(img)
```

4.Write text on an image

```python
import cv2
from google.colab.patches import cv2_imshow
from google.colab import files

img = cv2.imread(list(files.upload().keys())[0])

cv2.putText(img, "Hello", (50,100),
            cv2.FONT_HERSHEY_SIMPLEX, 1,
            (255,0,0), 2)

cv2_imshow(img)
```

5.Create a simple drawing canvas

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

canvas = np.zeros((300,300,3), dtype='uint8')

cv2.line(canvas,(0,0),(300,300),(255,255,255),2)
cv2.circle(canvas,(150,150),50,(0,255,0),2)

cv2_imshow(canvas)
```
