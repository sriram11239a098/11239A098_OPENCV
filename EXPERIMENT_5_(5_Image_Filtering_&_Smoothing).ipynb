1.Import OpenCV library.

```python
import cv2
from google.colab.patches import cv2_imshow
from google.colab import files
```

2.Read the input image

```python
uploaded = files.upload()
filename = list(uploaded.keys())[0]
img = cv2.imread(filename)
```

3.Apply different filters:

Averaging filter

Gaussian filter

Median filter

Bilateral filter

```python
avg = cv2.blur(img, (5,5))                  # Averaging filter
gauss = cv2.GaussianBlur(img, (5,5), 0)     # Gaussian filter
median = cv2.medianBlur(img, 5)             # Median filter
bilateral = cv2.bilateralFilter(img, 9, 75, 75)  # Bilateral filter
```

4.Display all images

```python
print("Original Image")
cv2_imshow(img)

print("Averaging Filter")
cv2_imshow(avg)

print("Gaussian Filter")
cv2_imshow(gauss)

print("Median Filter")
cv2_imshow(median)

print("Bilateral Filter")
cv2_imshow(bilateral)
```
