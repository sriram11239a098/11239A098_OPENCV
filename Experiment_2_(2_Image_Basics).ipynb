1.Display image properties (shape, size, datatype)

```python
import cv2
from google.colab import files

uploaded = files.upload()
filename = list(uploaded.keys())[0]

img = cv2.imread(filename)

print("Shape:", img.shape)
print("Size:", img.size)
print("Datatype:", img.dtype)
```

2.Color Space Conversion (RGB, Grayscale, HSV)

```python
import cv2
from google.colab.patches import cv2_imshow
from google.colab import files

uploaded = files.upload()
filename = list(uploaded.keys())[0]

img = cv2.imread(filename)

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

cv2_imshow(img)
cv2_imshow(gray)
cv2_imshow(hsv)
```

3.Split and merge color channels

```python
import cv2
from google.colab.patches import cv2_imshow
from google.colab import files

uploaded = files.upload()
filename = list(uploaded.keys())[0]

img = cv2.imread(filename)

b, g, r = cv2.split(img)

cv2_imshow(b)
cv2_imshow(g)
cv2_imshow(r)

merged = cv2.merge((b, g, r))
cv2_imshow(merged)
```

4.Extract Region of Interest (ROI)

```python
import cv2
from google.colab.patches import cv2_imshow
from google.colab import files

uploaded = files.upload()
filename = list(uploaded.keys())[0]

img = cv2.imread(filename)

roi = img[50:200, 50:200]   # (y1:y2, x1:x2)

cv2_imshow(roi)
```

5.Change image resolution and aspect ratio

```python
import cv2
from google.colab.patches import cv2_imshow
from google.colab import files

uploaded = files.upload()
filename = list(uploaded.keys())[0]

img = cv2.imread(filename)

resized = cv2.resize(img, (300, 300))

aspect = cv2.resize(img, (400, 200))

cv2_imshow(resized)
cv2_imshow(aspect)
```
